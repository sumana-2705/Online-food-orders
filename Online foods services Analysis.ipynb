# Food Data Analysis


```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns


print('hello world')
```

```python
df= pd.read_csv('onlinefoods.csv')

df.head()
```

```python
df.dtypes
```

```python
plt.style.use('fivethirtyeight')
```

## Finding missing Values

```python
df.isna().sum()
```

No missing Values found so we can proceed further for Analysis

## Data Visualisation

```python
fig,ax =plt.subplots(nrows=1,ncols=2)
g=sns.countplot(x=df['Occupation'],ax=ax[0])
g.set_xticklabels(g.get_xticklabels(), rotation=90)
g
plt.pie(df['Occupation'].value_counts(),labels=df['Occupation'].unique(),autopct='%0.1f%%')
```

Here we can observe that most of the customers are students and contribute 53.4% of total and rare customers are housewives who contributes less than 2.5%

```python
g=sns.catplot(data=df,x='Occupation',kind='count',col='Monthly Income',col_wrap=5)
g.set_xticklabels(rotation=90)
g
plt.show()
```

Here we can see more orders are from
----------students with no income
----------Employees with income range of 25k to 50k
----------Self Employed with income range more than 50k 
----------housewives with no income 

```python
g=sns.displot(kind='hist',data=df,x='Occupation',hue='Educational Qualifications')
g.set_xticklabels(rotation=90)
g
```

Among Students Postgraduates are in greater part and among Employees and Self employed graduates dominates

```python

g=sns.catplot(data=df,x='Occupation',kind='count',hue='Gender',col='Marital Status')
g.set_xticklabels(rotation=90)
g
```

Single Male students,Married Male employees and self employed Male are ordering more.Females are ordering less foods compared to males

```python
g=sns.displot(kind='hist',data=df,x='Occupation',hue='Feedback',col='Educational Qualifications')
g.set_xticklabels(rotation=90)
g
```

most of all categories are positive feedbacks, Employees who are graduate have given more negative feedbacks than others.

```python
g=sns.displot(kind='hist',data=df,x='Occupation',hue='Feedback',col='Gender')
g.set_xticklabels(rotation=90)
g
```

Here we can see a positive result that both male and female are giving more positive feedbacks,And as we from earlier graph more customers are male but here we can observe percentage of female is much lesser in Self employes part.

```python
fig,ax = plt.subplots(ncols=2,nrows=1)
g=sns.countplot(x=df['Marital Status'],hue=df['Feedback'],ax=ax[0])
g.set_xticklabels(g.get_xticklabels(), rotation=90)
g
plt.pie(df['Marital Status'].value_counts(),labels=df['Marital Status'].unique(),autopct='%0.1f%%')
```

Here we can observe most of the customers are single and have given very high positive feedbacks, positive feedbacks are also more among married couple but ratio is less comparitively with singles

```python
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(15,15))
sns.kdeplot(data=df,hue='Marital Status',x='Family size',ax=ax[0,0])
sns.barplot(data= df,y='Family size',x='Marital Status',hue='Feedback',ax=ax[0,1])
sns.barplot(data=df,y='Family size',x='Gender',hue='Feedback',ax=ax[1,0])
sns.kdeplot(data=df,hue='Gender',x='Family size',ax=ax[1,1])
```

1)However overall feedbacks are positive, If we analyse in depth who are giving more negative feedbacks,here we can observe Married people with family size '4' and people who dont preferred to say their marital status with family size between 2 to 3 have given more negative feedbacks

2)Females with family size in between 3 and 4 have given overall negative feedback

Married people with more than 50k income,single people (students) with no income are major part
